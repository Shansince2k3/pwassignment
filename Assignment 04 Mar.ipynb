### Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.graph_objects as go
import numpy as np
```

```python
titanic_data = sns.load_dataset("titanic")
num_points = 892
colors = np.random.choice(['red', 'green', 'blue', 'purple', 'orange'], num_points)
# Create a scatter plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=titanic_data.age, y=titanic_data.fare, mode='markers', marker=dict(color=colors)))

# Customize the layout (optional)
fig.update_layout(title='Scatter Plot Fare & Age',
                  xaxis_title='fare',
                  yaxis_title='Age')

# Show the plot
fig.show()
```

## Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.


```python
tips = sns.load_dataset("tips")
import plotly.express as px
# Load the tips dataset from plotly.express
tips_data = px.data.tips()
# Create a box plot using plotly.express
fig = px.box(tips_data, x='day', y='total_bill', color='sex', notched=True,
             title='Box Plot of Total Bill by Day and Gender',
             labels={'day': 'Day of Week', 'total_bill': 'Total Bill'},
             category_orders={'day': ['Thur', 'Fri', 'Sat', 'Sun']})
# Show the plot
fig.show()
```

## Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
fig = px.histogram(tips_data, x='sex', y='total_bill', color='day', pattern_shape='smoker',
      title='Histogram of Total Bill by Gender, Day, and Smoker Status',
      labels={'sex': 'Gender', 'total_bill': 'Total Bill'},
      category_orders={'day': ['Thur', 'Fri', 'Sat', 'Sun'], 'sex': ['Male', 'Female']})
fig.show()
```

## Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
### Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the dimensions parameter.

```python
# Load the iris dataset from plotly.express
iris_data = px.data.iris()

# Create a scatter matrix plot using plotly
fig = px.scatter_matrix(iris_data,
                        dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                        color="species",
                        symbol="species",
                        title="Scatter Matrix Plot of Iris Dataset",
                        opacity=0.7)
fig.update_layout(width=1400, height=1000)
# Show the plot
fig.show()
```

## Q5. What is Distplot? Using Plotly express, plot a distplot.

## Ans
### A distplot is a statistical plot that shows the distribution of a dataset. It is a combination of a histogram, a kernel density estimation (KDE) curve, and a rug plot. The histogram shows the frequency of data points in each bin, the KDE curve shows the smooth distribution of the data, and the rug plot shows the individual data points.
### Plotly Express is a high-level interface to Plotly that makes it easy to create interactive plots. To plot a distplot using Plotly Express, you can use the following code:

```python
import plotly.figure_factory as ff
import numpy as np
np.random.seed(2)

x = np.random.randn(1000)
hist_data = [x]
group_labels = ['distplot'] # name of the dataset

fig = ff.create_distplot(hist_data, group_labels)
fig.show()
```
