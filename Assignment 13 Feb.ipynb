# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

#### Ans: "LookupError" class exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.

```python
# KeyError: This error occurs when a dictionary key or a set element is not found.
my_dict = {"python": 1, "c": 2, "c++": 3}

try:
    count = my_dict["java"]
except KeyError as e:
    print("Error: Key not found!")
```

```python
# IndexError: This error occurs when a sequence index is out of range.

my_list = [1, 2, 3, 4, 5]

try:
    value = my_list[10]
except IndexError as e:
    print("Error: Index out of range!")
```

# Q6. List down some best practices for exception handling in python.

# ans
#### 1 Be specific: Catch only the exceptions you expect and can handle.
#### 2 Use separate blocks: Handle different exceptions separately.
#### 3 Utilize finally: Clean up resources in the finally block.
#### 4 Avoid unnecessary try-except: Only wrap code that can raise exceptions.
#### 5 Provide clear messages: Include helpful error messages.
#### 6 Handle exceptions at the right level: Let exceptions propagate to higher levels.
#### 7 Don't ignore exceptions: Handle exceptions appropriately.
#### 8 Use context managers: Simplify resource management with with statements.
#### 9 Log exceptions: Keep a record of errors for debugging.
#### 10 Test exception scenarios: Verify exception handling in your code.
