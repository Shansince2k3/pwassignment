# Q1. What is the Probability density function?

# Ans


The PDF is used for continuous probability distributions, such as the normal distribution, exponential distribution, and many others. It's important to note that for continuous variables, the probability of the variable taking on any specific value is technically zero. Instead, the PDF provides the probability of the variable falling within a small interval around that value.

The PDF is often denoted as $(f(x))$, where $(x)$ represents the value of the continuous random variable. The integral of the PDF over a specific interval gives the probability that the random variable falls within that interval.

For example, in the case of the normal distribution, the PDF is given by the bell-shaped curve formula, and it describes how likely different values are around the mean.

Mathematically, the probability $(P(a \leq X \leq b))$ that a continuous random variable \(X\) falls within the interval \([a, b]\) is calculated by integrating the PDF \(f(x)\) over that interval:

$[ P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx ]$

The PDF is a fundamental concept in continuous probability distributions, enabling us to analyze and make probabilistic statements about continuous random variables.

## The PDF has several important properties, including:  

- Non-negativity: f(x) is always non-negative, meaning that the probability of X taking on any value cannot be negative.
- Normalization: The area under the curve of f(x) must be equal to 1, since the probability of X taking on any value must be 1.
- Integration: The probability of X taking on a value within a certain range is given by the area under the curve of f(x) over that range.


## Q2. What are the types of Probability distribution?

### Ans
---

## There are many types of probability distributions, but some of the most commonly used distributions are:

1. **Normal Distribution (Gaussian Distribution)**:
   - Characteristics: Bell-shaped curve, symmetric, defined by mean (μ) and standard deviation (σ).
   - Mean: μ (mean of the distribution)
   - Variance: σ^2 (variance of the distribution)
   - Applications: Widely used in modeling real-world data such as measurements, IQ scores, and financial data.
   - Formula: f(x) = $(1 / (σ * √(2π))) * e^(-((x - μ)^2) / (2 * σ^2))$, where μ is the mean and σ is the standard deviation.
   - Implementation: Used for statistical inference, hypothesis testing, and generating random samples.

2. **Binomial Distribution**:
   - Characteristics: Discrete distribution, describes the number of successes in a fixed number of independent Bernoulli trials.
   - Mean: μ = n * p (mean of the distribution)
   - Variance: σ^2 = n * p * (1 - p) (variance of the distribution)
   - Formula: P(X = k) = $(n choose k) * p^k * (1 - p)^(n - k)$, where n is the number of trials, k is the number of successes, and p is the probability of success.
   - Applications: Modeling the number of successes in a series of binary events, such as coin flips or pass/fail scenarios.
   - Implementation: Used for analyzing experiments with binary outcomes, such as A/B testing.

3. **Poisson Distribution**:
   - Characteristics: Discrete distribution, models the number of events occurring in a fixed interval of time or space.
   - Mean: μ = λ (mean of the distribution)
   - Variance: σ^2 = λ (variance of the distribution)
   - Formula: P(X = k) = $(λ^k * e^(-λ)) / k!$, where λ is the average rate of occurrence and k is the number of events.
   - Applications: Used to model rare events, such as the number of customer arrivals at a store or the number of phone calls at a call center.
   - Implementation: Applied to situations involving counting events over a fixed interval.

4. **Exponential Distribution**:
   - Characteristics: Continuous distribution, models the time between events in a Poisson process.
   - Mean: μ = 1 / λ (mean of the distribution)
   - Variance: σ^2 = 1 / λ^2 (variance of the distribution)
   - Formula: $f(x) = λ * e^(-λx)$, where λ is the rate parameter
   - Applications: Used in reliability analysis, queueing theory, and survival analysis.
   - Implementation: Useful for modeling time-to-failure data and understanding waiting times.

5. **Uniform Distribution**:
   - Characteristics: Continuous or discrete distribution, all outcomes are equally likely.
   - Mean: μ = (a + b) / 2 (mean of the distribution, for continuous uniform)
   - Variance: $σ^2 = (b - a)^2 / 12$ (variance of the distribution, for continuous uniform)
   - Applications: Often used when each outcome in a sample space is equally likely, such as in random number generation or modeling uncertainty.
   - Implementation: Commonly used in simulations and Monte Carlo methods.

## Q3. Write a Python function to calculate the probability density function of a normal distribution with  given mean and standard deviation at a given point.

```python
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    """
    exponent = -(x - mean)**2 / (2 * std_dev**2)
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    pdf_value = coefficient * math.exp(exponent)
    return pdf_value

# Example usage
mean = 0
std_dev = 1
x_point = 0.5
pdf_value_at_x = normal_pdf(x_point, mean, std_dev)
print(f"PDF at x = {x_point}: {pdf_value_at_x}")
```

## Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

## Ans 
_______
### The Binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials. It has several properties that make it useful for modeling various scenarios:

### Properties of the Binomial Distribution:

- Fixed Number of Trials: The distribution is defined for a fixed number of trials, denoted as 'n', where each trial can have one of two outcomes: success or failure.
- Independent Trials: Each trial is assumed to be independent of the others, meaning the outcome of one trial does not affect the outcome of subsequent trials.
- Two Outcomes: Each trial results in one of two possible outcomes, often labeled as success (S) and failure (F).
- Constant Probability: The probability of success, denoted as 'p', remains constant across all trials. The probability of failure is then '1 - p'.
- Discrete Distribution: The Binomial distribution is a discrete distribution since it deals with a count of successes, which is always a whole number.
- Formula: The probability of observing exactly 'k' successes in 'n' trials is given by the formula: P(X = k) =  $ (n \choose k) * p^k * (1 - p)^{n - k}$, where 'n choose k' represents the number of ways to choose 'k' successes from 'n' trials.
- Mean and Variance: The mean (μ) of the Binomial distribution is μ = n * p, and the variance (σ^2) is σ^2 = n * p * (1 - p).

## EXAMPLE
 

1. Flipping a coin: If you flip a fair coin (with an equal probability of heads or tails) 10 times, you can use the binomial distribution to calculate the probability of getting a certain number of heads (or tails) out of those 10 flips.

2. Quality control: A company produces light bulbs, and they want to know the probability that a batch of 100 bulbs will have more than 10 defective bulbs. They can use the binomial distribution to model the probability of each bulb being defective, and then use that to calculate the probability of getting more than 10 defective bulbs in the batch.


## Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate example data with a normal distribution
mu = 1000  
sigma = 0.4 
data = np.random.normal(mu, sigma, 1000)  # Simulated height data


# Calculate the PDF of the normal distribution
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = norm.pdf(x, mu, sigma)

# Create subplots
fig, ax1 = plt.subplots()

# Plot the histogram of the data
ax1.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Histogram')

# Plot the PDF as a curve
ax2 = ax1.twinx()
ax2.plot(x, pdf, color='r', label='PDF (Normal Distribution)')

# Set x-axis ticks to show standard deviations and data percentages
xticks = [mu - 3*sigma, mu - 2*sigma, mu - sigma, mu, mu + sigma, mu + 2*sigma, mu + 3*sigma]
xticklabels = ['-3σ\n0.13%', '-2σ\n2.28%', '-σ\n15.87%', 'μ\n34.13%', 'σ\n15.87%', '2σ\n2.28%', '3σ\n0.13%']
ax1.set_xticks(xticks)
ax1.set_xticklabels(xticklabels)

# Annotate data percentages within each standard deviation range
for i in range(len(xticks)):
    ax1.annotate(xticklabels[i], (xticks[i], 0), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='blue')

# Set labels and title
ax1.set_xlabel('Height')
ax1.set_ylabel('Density (Histogram)')
ax2.set_ylabel('Density (PDF)')
plt.title('Height Distribution, Histogram, and Probability Density Function')

# Show legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=0)

plt.grid(True)
plt.show()
```

## Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

```python
import math

def poisson_cdf(x, mean):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution.
    """
    cdf_value = 0
    for k in range(math.floor(x) + 1):
        cdf_value += (math.exp(-mean) * mean**k) / math.factorial(k)
    return cdf_value

# Example usage
mean = 3
x_point = 2
cdf_value_at_x = poisson_cdf(x_point, mean)
print(f"CDF at x = {x_point}: {cdf_value_at_x}")
```

## Q7. How Binomial distribution different from Poisson distribution?

## Ans
-----

| **Aspect**                       | **Binomial Distribution**                                   | **Poisson Distribution**                                       |
|----------------------------------|------------------------------------------------------------|---------------------------------------------------------------|
| Nature of Events                 | Binary outcomes (success/failure) in fixed trials.        | Rare events in fixed interval of time/space.                |
| Number of Trials/Time Intervals  | Fixed number of trials ('n').                              | No fixed number of trials; focuses on fixed interval.      |
| Number of Possible Outcomes      | Discrete count of successes in 'n' trials.               | Discrete count of rare events.                             |
| Probability Parameter            | 'n' (number of trials) and 'p' (probability of success). | 'λ' (average rate of event occurrence).                   |
| PDF                              | $P(X = k) = (n \choose k) * p^k * (1 - p)^{n - k}$        | $P(X = k) = \frac{λ^k * e^{-λ}}{k!}$                       |
| CDF                              | Calculated using PDF or cumulative sum of PMF.            | Calculated using summation of PMF.                        |
| Shape                            | Symmetrical if 'p' is close to 0.5; skewed if not.        | Right-skewed, as 'λ' measures average event rate.         |
| Mean                             | μ = n * p                                               | μ = λ                                                     |
| Variance                         | σ^2 = n * p * (1 - p)                                  | σ^2 = λ                                                     |
| Assumptions                      | Trials are independent and have constant 'p'.           | Events occur independently and at a constant rate.       |
| Applicability                    | Quality control, A/B testing, multiple-choice questions. | Modeling rare events (accidents, diseases, disasters).   |
| Limiting Cases                   | Approaches Poisson with large 'n', small 'p'.            | N/A.                                                      |


## Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np

# Parameters for the Poisson distribution
mean = 5
sample_size = 1000

# Generate the random sample
random_sample = np.random.poisson(mean, size=sample_size)

# Calculate the sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

## Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

## Ans
----

| **Distribution**       | **Mean (μ)**     | **Variance (σ^2)**                     |
|-----------------------|------------------|----------------------------------------|
| **Binomial**          | μ = n * p        | σ^2 = n * p * (1 - p)                  |
| **Poisson**           | μ = λ            | σ^2 = λ                                |

In the Binomial distribution, the variance is influenced by both the number of trials 'n' and the probability of success 'p'. In the Poisson distribution, the variance is solely determined by the average rate 'λ'. Furthermore, in the Poisson distribution, the mean and variance are equal, reflecting the distribution's property that rare events occur with a constant average rate.


## Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

## Ans
--------
-  the data is symmetrically distributed around the mean. This means that the least frequent data appears in the tails of the distribution, farthest from the mean. The tails of a normal distribution are the regions that extend to the left and right beyond the central peak (where the mean is located).

- In a standard normal distribution (with a mean of 0 and a standard deviation of 1), the data becomes progressively less frequent as you move away from the mean in either direction. This is known as the "bell-shaped" curve, and it reflects the typical behavior of data in many natural phenomena and measurements.

- So, the least frequent data appears in the tails of the normal distribution, which are located at the extreme left and right sides, away from the central peak or mean position.
