# Q1. What are the three measures of central tendency?

### Ans :
The three measures of central tendency are:

1. Mean: The mean is the sum of all values in a dataset divided by the total number of values. It represents the average value of a dataset.

2. Median: The median is the middle value of a dataset when the values are arranged in ascending order. If there are an even number of values, the median is the average of the two middle values.

3. Mode: The mode is the value that occurs most frequently in a dataset. If there are multiple values with the same highest frequency, the dataset is said to have multiple modes.

## Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset? Provide an example.

### Ans

| Measure           | Definition and Calculation                     | Interpretation                            | Example                                 |
|-------------------|------------------------------------------------|-------------------------------------------|-----------------------------------------|
| Range             | Difference between max and min values          | Indicates total spread                   | Data: [10, 15, 20, 25, 30]             |
|                   | Range = Max value - Min value                  | Range = 30 - 10 = 20                      | Range = 20                             |
| Variance          | Average of squared differences from the mean   | Measures average deviation from the mean | Mean = (10 + 15 + 20 + 25 + 30) / 5 = 20 |
|                   | Square differences: (x - mean)^2               | Variance ≈ (100 + 25 + 0 + 25 + 100) / 5 ≈ 50 |
|                   | Variance = Σ(squared differences) / n         | Variance = 50                            |
| Standard Deviation| Square root of the variance                   | Provides spread in original units         | Standard Deviation ≈ √50 ≈ 7.07       |
|                   | Standard Deviation = √Variance                 | Standard Deviation ≈ 7.07                |


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
from scipy import stats

# Generate example data
np.random.seed(42)
data = np.random.normal(50, 10, 1000)  # Generating a normal distribution

# Calculate mean, mode, and median
mean = np.mean(data)
with warnings.catch_warnings():  # Suppress the FutureWarning
    warnings.simplefilter("ignore")
    mode = stats.mode(data)[0][0]
median = np.median(data)



# Create a displot
sns.set(style="whitegrid")
sns.displot(data, kde=True)

# Mark the mean, mode, and median on the plot
plt.axvline(mean, color='r', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axvline(mode, color='g', linestyle='--', label=f'Mode: {mode:.2f}')
plt.axvline(median, color='b', linestyle='--', label=f'Median: {median:.2f}')

plt.legend()
plt.title('Distribution with Mean, Mode, and Median')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

## Q8. What do you understand about skewness in data?

## ANS


There are two main types of skewness

| Aspect                                      | 1. Positively Skewed Data                | 2. Negatively Skewed Data               |
|---------------------------------------------|--------------------------------------|-------------------------------------|
| **Mean vs. Median**                         | Mean > Median                        | Mean < Median                       |
| **Tail Direction**                         | Longer/fatter tail on the right     | Longer/fatter tail on the left      |
| **Data Distribution**                      | Majority of data on the left        | Majority of data on the right       |
| **Impact on Central Tendency Measures**    | Mean is pulled towards the tail     | Mean is pulled towards the tail     |
| **Impact on Spread Measures**              | Greater range on the right side     | Greater range on the left side      |
| **Data Interpretation**                    | May indicate potential outliers     | May indicate potential outliers     |
| **Decision Making**                        | May overestimate central tendency   | May underestimate central tendency   |
| **Statistical Analyses**                   | May affect assumptions of normality | May affect assumptions of normality |


#### Skewness in data refers to the asymmetry or lack of symmetry in the distribution of values. In a symmetrical distribution, the data is evenly distributed around a central point, and the left and right sides are roughly mirror images of each other. Skewness occurs when this symmetry is disrupted, causing the distribution to be shifted to one side.

## Q9. If a data is right skewed then what will be the position of median with respect to mean?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate example right-skewed data
np.random.seed(42)
data = np.random.exponential(scale=2, size=1000)  # Generating right-skewed data

# Calculate mean and median
mean = np.mean(data)
median = np.median(data)

# Create a KDE plot
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
sns.kdeplot(data, fill=True, color='blue', label='Data Distribution')  # Use fill instead of shade

# Mark the mean and median on the plot
plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='--', label=f'Median: {median:.2f}')

plt.legend()
plt.title('Right-Skewed Data Distribution with Mean and Median')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

## Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?

## Ans
#### Correlation is a more commonly used measure than covariance, because it is standardized and easier to interpret. Correlation can also be used to compare the strength of the relationship between different pairs of variables on a common scale.

## ANS

|

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate example data
np.random.seed(42)
x = np.random.randint(0, 20, 50)  # Random x values
y = x + np.random.normal(0, 3, 50)  # Y values with some noise

# Calculate covariance and correlation
covariance = np.cov(x, y)[0, 1]
correlation = np.corrcoef(x, y)[0, 1]

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y, color='blue', label='Data Points')
plt.title('Scatter Plot with Covariance and Correlation')
plt.xlabel('X')
plt.ylabel('Y')

# Annotate covariance and correlation values
plt.annotate(f'Covariance: {covariance:.2f}', (5, 30), color='red')
plt.annotate(f'Correlation: {correlation:.2f}', (5, 27), color='green')

plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate example data
np.random.seed(42)
data = np.random.randint(0, 20, (10, 5))  # Generating example data

# Calculate covariance and correlation matrices
cov_matrix = np.cov(data, rowvar=False)  # Rowvar=False for variables in columns
corr_matrix = np.corrcoef(data, rowvar=False)

# Set up subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
sns.set()

# Heatmap for Covariance Matrix
sns.heatmap(cov_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            xticklabels=['Var1', 'Var2', 'Var3', 'Var4', 'Var5'],
            yticklabels=['Var1', 'Var2', 'Var3', 'Var4', 'Var5'],
            ax=axes[0])
axes[0].set_title("Covariance Matrix")

# Heatmap for Correlation Matrix
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            xticklabels=['Var1', 'Var2', 'Var3', 'Var4', 'Var5'],
            yticklabels=['Var1', 'Var2', 'Var3', 'Var4', 'Var5'],
            ax=axes[1])
axes[1].set_title("Correlation Matrix")

plt.tight_layout()
plt.show()
```

## Q11. What is the formula for calculating the sample mean? Provide an example calculation for a dataset.

## Calculating Sample Mean

The formula for calculating the sample mean (also known as the average) is:

Sample Mean = Sum of all data points / Number of data points

Mathematically:

Sample Mean= $\frac{\sum_{i=1}^{n} x_i}{n}$

Where:
- xi represents each individual data point.
- n is the total number of data points.

### Example Calculation:

Consider the dataset: [10, 15, 20, 25, 30]

To calculate the sample mean:

Sample Mean = (10 + 15 + 20 + 25 + 30) / 5 = 100 / 5 = 20

So, the sample mean of the dataset is 20.

## Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

## Ans 
Outliers are extreme values that deviate significantly from the rest of the data points in a dataset. They can have a significant impact on measures of central tendency (mean, median, mode) and measures of dispersion (range, variance, standard deviation).

```python
import numpy as np

# Generate example data with an outlier
data_with_outlier = np.array([85, 90, 92, 88, 95, 200])

# Calculate measures without outlier
mean_without_outlier = np.mean(data_with_outlier[:-1])  # Excluding the outlier
median_without_outlier = np.median(data_with_outlier[:-1])
mode_without_outlier = np.nan  # No mode in this case
range_without_outlier = np.ptp(data_with_outlier[:-1])  # Peak-to-Peak (range)
variance_without_outlier = np.var(data_with_outlier[:-1])
std_deviation_without_outlier = np.std(data_with_outlier[:-1])

# Calculate measures with outlier
mean_with_outlier = np.mean(data_with_outlier)
median_with_outlier = np.median(data_with_outlier)
mode_with_outlier = np.nan  # No mode in this case
range_with_outlier = np.ptp(data_with_outlier)
variance_with_outlier = np.var(data_with_outlier)
std_deviation_with_outlier = np.std(data_with_outlier)

print("Measures without outlier:")
print("Mean:", mean_without_outlier)
print("Median:", median_without_outlier)
print("Range:", range_without_outlier)
print("Variance:", variance_without_outlier)
print("Standard Deviation:", std_deviation_without_outlier)

print("\nMeasures with outlier:")
print("Mean:", mean_with_outlier)
print("Median:", median_with_outlier)
print("Range:", range_with_outlier)
print("Variance:", variance_with_outlier)
print("Standard Deviation:", std_deviation_with_outlier)
```
