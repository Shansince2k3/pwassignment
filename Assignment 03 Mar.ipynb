## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

### 1. Scatter Plot:
   - Uses: Scatter plots display the relationship between two continuous variables. They are useful for visualizing patterns or correlations in data.
### 2. Line Plot:
   - Uses: Line plots show the trend or evolution of a continuous variable over time or another continuous variable. They are suitable for time series analysis or plotting trends.
### 3. Bar Plot:
   - Uses: Bar plots represent categorical data with rectangular bars. They are helpful for comparing values across different categories or groups.
### 4. Histogram:
   - Uses: Histograms display the distribution of a continuous variable by dividing it into bins and counting the frequency of observations in each bin. They provide insights into the data's shape, central tendency, and spread.
### 5. Heatmap:
   - Uses: Heatmaps visualize data using a color-coded matrix, where each cell represents the relationship between two categorical variables or the correlation between two continuous variables. They are valuable for identifying patterns, clusters, or relationships in data.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 7, 8]

# Creating a scatter plot
sns.scatterplot(x=x, y=y)
plt.show()
sns.lineplot(x=x, y=y)
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 12, 8]

# Creating a bar plot
sns.barplot(x=categories, y=values)
plt.show()
```

```python
data = [1, 1, 2, 2, 2, 3, 3, 4, 5]
# Creating a histogram
sns.histplot(data=data, bins=5)
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = np.random.rand(5, 5)

# Creating a heatmap
sns.heatmap(data, annot=True, cmap='YlGnBu')
plt.show()
```

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")
# Create the line plot
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Set plot title and labels
plt.title("fMRI Signal over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Display the plot
plt.show()
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'. Note: pclass, age, and fare are columns in the titanic dataset.

```python

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create the box plots
fig,axes = plt.subplots(1, 2, figsize=(10, 5))

# Box plot for 'age'
sns.boxplot(ax=axes[0], data=titanic_data, x='pclass', y='age')
axes[0].set_title('Age by Passenger Class')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Age')

# Box plot for 'fare'
sns.boxplot(ax=axes[1], data=titanic_data, x='pclass', y='fare')
axes[1].set_title('Fare by Passenger Class')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Fare')

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")
# Plot histogram with hue
sns.histplot(data=diamonds_data, x="price", hue="cut", multiple="stack")
# Set plot title and labels
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")
# Display the plot
plt.show()
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris_data = sns.load_dataset("iris")
# Plot pair plot with hue
sns.pairplot(data=iris_data, hue="species")
# Display the plot
plt.show()
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights_data = sns.load_dataset("flights")
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Create the heatmap
sns.heatmap(data=flights_pivot, cmap="YlGnBu")

# Set plot title and labels
plt.title("Number of Passengers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")

# Display the plot
plt.show()
```
